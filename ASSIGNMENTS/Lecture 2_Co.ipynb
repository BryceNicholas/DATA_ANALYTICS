```python
from ucimlrepo import fetch_ucirepo
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

heart_failure_clinical_records = fetch_ucirepo(id=519)

X = heart_failure_clinical_records.data.features
y = heart_failure_clinical_records.data.targets
X['target'] = y

# Distribution of Age
plt.figure(figsize=(8, 6))
sns.histplot(X['age'], bins=20, kde=True, color='skyblue')
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

# Distribution of Gender
plt.figure(figsize=(6, 4))
sns.countplot(x='sex', data=X, palette='Set2')
plt.title('Distribution of Gender')
plt.xlabel('Gender (0 = Female, 1 = Male)')
plt.ylabel('Count')
plt.show()

# Distribution of Anaemia
plt.figure(figsize=(6, 4))
sns.countplot(x='anaemia', data=X, palette='Set1')
plt.title('Distribution of Anaemia')
plt.xlabel('Anaemia (0 = No, 1 = Yes)')
plt.ylabel('Count')
plt.show()

# Distribution of High Blood Pressure
plt.figure(figsize=(6, 4))
sns.countplot(x='high_blood_pressure', data=X, palette='Set3')
plt.title('Distribution of High Blood Pressure')
plt.xlabel('High Blood Pressure (0 = No, 1 = Yes)')
plt.ylabel('Count')
plt.show()

# Distribution of Death Event (target variable)
plt.figure(figsize=(6, 4))
sns.countplot(x='target', data=X, palette='Set1')
plt.title('Distribution of Death Event')
plt.xlabel('Death Event (0 = No, 1 = Yes)')
plt.ylabel('Count')
plt.show()

# Relationship between Age and Death Event
plt.figure(figsize=(8, 6))
sns.boxplot(x='target', y='age', data=X, palette='Set2')
plt.title('Age vs Death Event')
plt.xlabel('Death Event (0 = No, 1 = Yes)')
plt.ylabel('Age')
plt.show()

# Relationship between Ejection Fraction and Death Event
plt.figure(figsize=(8, 6))
sns.boxplot(x='target', y='ejection_fraction', data=X, palette='Set2')
plt.title('Ejection Fraction vs Death Event')
plt.xlabel('Death Event (0 = No, 1 = Yes)')
plt.ylabel('Ejection Fraction')
plt.show()

# Relationship between Serum Creatinine and Death Event
plt.figure(figsize=(8, 6))
sns.boxplot(x='target', y='serum_creatinine', data=X, palette='Set2')
plt.title('Serum Creatinine vs Death Event')
plt.xlabel('Death Event (0 = No, 1 = Yes)')
plt.ylabel('Serum Creatinine')
plt.show()

# Create a pairplot to visualize multiple relationships
sns.pairplot(X, hue='target', vars=['age', 'ejection_fraction', 'serum_creatinine', 'platelets'], palette='Set1')
plt.suptitle('Pairplot of Various Features Colored by Death Event', y=1.02)
plt.show()
```
